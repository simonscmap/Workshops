## Version 0.2.14+


```python
# !pip install pycmap -U -q    

import pycmap
api = pycmap.API()
pycmap.__version__
```

## [*datasets_with_ancillary()*](https://github.com/simonscmap/pycmap/blob/master/docs/DatasetsWithAncillary.ipynb)

Returns a dataframe containing the list of data sets that have been automatically colocalized with ancillary variables (mostly environmental).
A growing number of Simons CMAP datasets are being automatically colocalized with a large (100+) number of ancillary parameters derived from satellite and numerical model products.

>**Returns:** 
>>  Pandas dataframe.


```python
api.datasets_with_ancillary()
```

## [*get_dataset_with_ancillary(tableName)*](https://github.com/simonscmap/pycmap/blob/master/docs/RetrieveDatasetWithAncillary.ipynb)

Returns the entire dataset joined with colocalized ancillary variables. The ancillary variable names are prefixed with `CMAP_`.
<br/>Note that this method does not return the dataset metadata. Use the get_dataset_metadata method to get the dataset metadata.




> **Parameters:** 
>> **tableName: string**
>>  <br />The name of table associated with the dataset. A full list of table names can be found in the [Data Catalog](https://simonscmap.com/catalog).
>> <br />

>**Returns:** 
>>  Pandas dataframe.



### Example:
Returns the unit of the short name variable, silica_hot, in the HOT Particle Flux dataset.

```python
df = api.get_dataset_with_ancillary('tblKM1709_mesoscope')
df
```

```python
df.to_csv('withAncillary.csv', index=False)
```
